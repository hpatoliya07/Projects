```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import openpyxl
```

```python
link="https://www.imdb.com/chart/top-english-movies/"
```

```python
page=requests.get(link)
```

```python
soup=bs(page.content, 'html.parser')
```

```python
movies=soup.find('tbody', class_="lister-list").find_all('tr')
```

```python
excel = openpyxl.Workbook()
print(excel.sheetnames)
sheet = excel.active
sheet.title = 'Top Rated Movies'
print(excel.sheetnames)
sheet.append(['Movie Rank', 'Movie Name', 'Year of Release', 'TMDB Rating'])


for movie in movies:
  rank = movie.find('td', class_='titleColumn').get_text(strip=True).split(".")[0]
  name = movie.find('td', class_='titleColumn').a.text
  release = movie.find('span', class_="secondaryInfo").text
  rating = movie.find('td', class_='ratingColumn imdbRating').text

  print(rank, name, release, rating)

  sheet.append([rank, name, release, rating]) 

excel.save('IMDB Movie Rating.xlsx')
```
